```python
import pandas as pd
Location = "C:/Users/liamk/Desktop/USF/BUSINESS INTELLIGENCE/datasets/gradedata.csv"
df = pd.read_csv(Location)
def score_to_numeric(x):
    if x=='female':
        return 1
    if x=='male':
        return 0
df['gender_val'] = df['gender'].apply(score_to_numeric) 
import statsmodels.formula.api as sm
result = sm.ols(
 formula='grade ~ exercise + hours + gender',
 data=df).fit()
result.summary()
```
